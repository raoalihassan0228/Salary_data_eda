# Salary Data Insights: A Detailed Exploratory Analysis

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading dataset
df = pd.read_csv('salary_data.csv')
df.head()
```

```python
# adding new column of 'age_group'
df['age_group']=pd.cut(df['Age'], [10,18,30,50,80],
                       labels=['child','youth','middle aged','old'])
```

### Filling null values of all columns

```python
# filling nan values 'age' column
df.Age.fillna(df['Age'].mean(), inplace=True)
```

```python
# filling nan values 'Years of Experience' column
df['Years of Experience'].fillna(df['Years of Experience'].mode()[0], inplace=True)
```

```python
# filling nan values 'salary' column
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```

```python
# filling nan values 'Gender' column
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
```

```python
# filling nan values 'Education Level' column
df['Education Level'].fillna(df['Education Level'].mode()[0], inplace=True)
```

```python
# filling nan values 'Job Title' column
df['Job Title'].fillna(df['Job Title'].mode()[0], inplace=True)
```

```python
df.isnull().sum()   # checking nan values
```

```python
df.describe()
```

```python
# boxplot for checking outliers
sns.boxplot(df, y='Salary')
```

```python
# barplot b/w Salary & Education Level
sns.barplot(data=df, y='Salary', x='Education Level', hue='Gender')
```

```python
# data distribution
sns.histplot(data=df, x='Salary',bins=20)
```

```python
# scatterplot
sns.scatterplot(df, x='Age', y='Salary')
```

```python
sns.scatterplot(df, x='Years of Experience', y='Salary')
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby('age_group')['Salary'].mean().plot(kind='bar')
```

```python
sns.pairplot(df)
```

```python
# pairplot for complete overview
sns.pairplot(df ,vars=['Age','Gender','Education Level','Years of Experience','Salary'])
```

```python
#stacked barchart/plot
df.groupby(['Years of Experience','Gender'])['Salary'].mean().unstack().plot(kind='bar',stacked=True)
```

```python
# swarm plot
sns.swarmplot(data=df, x='Education Level', y='Age', hue='Gender') #swarm plot gives more information
```

```python
# Pivot table (City, Cases)
table=pd.pivot_table(df, values= 'Salary', index='Education Level',columns='Gender',aggfunc=np.mean)
table
```

```python
# Select numerical columns for correlation
numerical_columns = ["Age", "Years of Experience", "Salary"]

# Create a correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Set the figure size (optional)
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Show the plot
plt.show()
```

```python
# Pivot table (Salary, Job Title, 'Education Level')
table=pd.pivot_table(df, values= 'Salary', index='Job Title',columns="Education Level",aggfunc=np.sum)
table.fillna(0, inplace = True)
pivot=table.style.background_gradient(cmap='icefire')
pivot
```
